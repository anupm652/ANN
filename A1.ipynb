```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1/1+(np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tan(x)

def leaky_relu(x, alpha = 0.01):
    return np.maximum(alpha*x, x)


x = np.linspace(-10, 10, 100)

def plot_activation_fun(x, value, label):

    plt.figure(figsize=(10, 6))
    plt.plot(x, value, label=label)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Activation Funcion:{label}')
    plt.legend()
    plt.show()
```

```python
sigmoid_y = sigmoid(x)
plot_activation_fun(x, sigmoid_y, label='Sigmoid')
```

```python
relu_y = relu(x)
plot_activation_fun (x, relu_y, label='ReLU')
```

```python
tanh_y = tanh(x)
plot_activation_fun(x, tanh_y, label='Tanh')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / 1 + np.exp(-x)

def ReLU(x):
    return np.maximum(0, x)

def Tanh(x):
    return np.tanh(x)

def L_ReLU(x):
    return np.maximum(0.01 * x, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x)
```

```python
x = np.linspace(-10, 10, 100)
```

```python
def plot_acti(x, value, label):
    plt.figure(figsize=(10, 6))
    plt.plot(x, value, label=label)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Activation Function : {label} ')
    plt.legend()
    plt.show()
```

```python
softmax_y = softmax(x)
plot_acti(x, softmax_y, label='softmax')
```

```python
sigmoid_y = sigmoid(x)
plot_acti(x, sigmoid_y, label='Sigmoid')
```
